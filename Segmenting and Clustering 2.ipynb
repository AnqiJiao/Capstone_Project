#### import libraries

```python
import pandas as pd
import numpy as np
```

#### import data

```python
# import toronto data from the first part
df_toronto = pd.read_csv("C:\\Users\\jiaoa\\Dropbox\\Data Science\\Python\\df_toronto.csv")
Postcode = df_toronto['Postcode']

# import data for geographical coordinates of each postal code
df_geo = pd.read_csv("C:\\Users\\jiaoa\\Dropbox\\Data Science\\Python\\Geospatial_Coordinates.csv")
df_geo.head()
```

#### merge and construct data with latitude and longitude

```python
# merge two data sets above
df_new = pd.merge(df_toronto, df_geo, how='inner', left_on=['Postcode'], right_on = ['Postal Code'])

# keep useful variables
df_new = df_new[['Postcode','Borough','Neighborhood','Latitude','Longitude']]

# export to csv file
df_new.to_csv('toronto_ll.csv')
df_new.head(11)
```
